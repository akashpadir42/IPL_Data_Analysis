### Fixing season column missing values 

Since season is a categorical column so for this we are just droping missing values

```python
data = data.dropna(subset=['season']).reset_index(drop=True)
```

### Fixing winner column missing values

Since winner column is a categorical column so for this we are just setting No Player for missing values

```python
data['winner'].fillna('No Player',inplace=True)
```

### Fixing pom column missing values

Since pom column is a categorical column so for this we are just setting No Player for missing values

```python
data['pom'].fillna('No Player',inplace=True)
```

### Fixing home_overs, home_runs, home_wickets, home_boundaries column missing values 

Since these emply cells are in same row because to no match conducted so for this we are just setting missing values to 0

```python
data['home_overs'].fillna(0,inplace=True)
data['home_runs'].fillna(0,inplace=True)
data['home_wickets'].fillna(0,inplace=True)
data['home_boundaries'].fillna(0,inplace=True)
```

### Fixing away_overs, away_runs, away_wickets, away_boundaries column missing values

Since these emply cells are in same row because to no match conducted so for this we are just setting missing values to 0

```python
data['away_overs'].fillna(0,inplace=True)
data['away_runs'].fillna(0,inplace=True)
data['away_wickets'].fillna(0,inplace=True)
data['away_boundaries'].fillna(0,inplace=True)
```

## Checking for missing values in the dataset.

```python
data.isna().sum()
```

## Changing Data type of columns

```python
data['start_date'] = pd.to_datetime(data['start_date'])
data['super_over'] = data['super_over'].astype(bool)
data['home_runs'] = data['home_runs'].astype('int16')
data['home_wickets'] = data['home_wickets'].astype('int16')
data['home_boundaries'] = data['home_boundaries'].astype('int16')
data['away_runs'] = data['away_runs'].astype('int16')
data['away_wickets'] = data['away_wickets'].astype('int16')
data['away_boundaries'] = data['away_boundaries'].astype('int16')
```

### Creating city column using venue_name column

```python
data['city'] = ''
for i in range(data.shape[0]):
    data.loc[i, 'city'] = data['venue_name'].str.split(', ')[i][-1]
```

### Rearranging columns

```python
data = data[['season', 'id', 'name', 'short_name', 'home_team', 'away_team',
       'toss_won', 'decision', 'winner', 'start_date', 'venue_id',
       'venue_name','city', 'home_captain', 'away_captain', 'pom', 'super_over',
       'home_overs', 'home_runs', 'home_wickets', 'home_boundaries',
       'away_overs', 'away_runs', 'away_wickets', 'away_boundaries']]
```

# Some Questions And Anwers

### Q1: What is the percentage of winning a match if a team wins the toss?

```python
success = data[data['toss_won'] == data['winner']].shape[0]
failure = data[data['toss_won'] != data['winner']].shape[0]

total_matches = success + failure

winning_perc = round(success/total_matches*100,2)
failure_perc = round(failure/total_matches*100,2)

print('Winning percentage:', winning_perc)
print('Loosing percentage:', failure_perc)
```

```python
plt.bar(['Win','Loss'], [winning_perc,failure_perc],width=0.8)
plt.ylabel("Percentage (in %)")
plt.title("Percentage of Winning or Loosing Match After Winning Toss")
plt.show()
```

This graph represents the winning and losing percentages of a team after winning a toss. Based on the graph, we can observe the following:
1. It shows that the winning and losing percentages are approximately equal.
2. This suggests that the outcome of the match after winning the toss is largely dependent on luck.
3. However, the winning percentage is slightly higher than the losing percentage.



### Q2: What is the total number of boundaries by the home team and away team per season?

```python
boundries = data.groupby('season')[['home_boundaries','away_boundaries']].sum()
boundries
```

```python
sns.lineplot(data=boundries)
plt.title('Boundaries by Teams in each Season')
plt.xlabel('Seasons')
plt.ylabel('Number of Boundaries')
plt.show()
```

This graph represents the number of boundaries made by the home team and away team per season. 
Based on this graph, we can observe the following:
- There is no significant disparity between the number of boundaries made by the home team and the away team.
- In each season, both teams have made nearly the same number of boundaries.
- The season with the fewest number of boundaries made by both teams was 2009.
- The away team made the highest number of boundaries in the 2022 season, while the home team made the highest number in the 2023 season.

### Q3: How many super overs occurred per season?

```python
super_overs = data.groupby('season')['super_over'].sum()
super_df = super_overs.reset_index()
ans = (super_df['super_over'] !=0).sum()
print('Total super overs occured in entire cricket period =',ans)
```

```python
plt.plot(super_overs)
plt.title('No. of super over occured per Season')
plt.xlabel('Seasons')
plt.ylabel('Number of super overs')
plt.show()
```

This graph represents the number of super overs that occurred per season. 
Based on the graph, we can conclude the following:

- A total of 7 super overs have occurred in the seasons from 2008 to 2023.
- The seasons with 0 super overs are 2008, 2009, 2010, 2011, 2012, 2016, 2018, 2022, and 2023.
- The highest number of super overs occurred in the 2020 season, with a total of 4 super overs.

### Q6. What is the average score of each stadium?

```python
venue_avg_scores = data.groupby('venue_name')['home_runs'].mean()
sorted_venues = venue_avg_scores.sort_values(ascending=False)
sorted_venues
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sorted_venues.index, sorted_venues.values, color='blue')
plt.title('Venue Analysis')
plt.xlabel('Venue')
plt.ylabel('Average Scores')
plt.xticks(rotation=90)
plt.show()
```

From this graph, we can observe the following:
1. The average at Barsapara Cricket Stadium, Guwahati is high, at almost 195 runs.
2. The average at Newlands, Cape Town is the lowest, at almost 111 runs.
3. Most stadiums have an average number of runs per inning falling within the range of 140 to 170 runs.

### Q7. What is the average wickets of each stadium?

```python
venue_avg_wkts = data.groupby('venue_name')['home_wickets'].mean()
sorted_venues = venue_avg_wkts.sort_values(ascending=False)
sorted_venues
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sorted_venues.index, sorted_venues.values, color='blue')
plt.title('Venue Analysis')
plt.xlabel('Venue')
plt.ylabel('Average Wkts')
plt.xticks(rotation=90)
plt.show()
```

From this graph, we can observe the following:

1. The average number of wickets per inning at Kingsmead, Durban, and Dr DY Patil Sports Academy, Navi Mumbai, is high, at approximately 7 wickets per inning.
2. The average number of wickets per inning at Buffalo Park, East London, is the lowest, at nearly 4 wickets per inning.
3. The majority of stadiums have an average number of wickets per inning ranging from 4.5 to 6 wickets.

### Q8.What is the win percentage of the team in each season?

```python
win_percentage = data.groupby(['season', 'home_team']).apply(lambda x: (x['home_team'] == x['winner']).mean()).unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(win_percentage, cmap='YlGnBu', annot=True, fmt='.2f', cbar=True)
plt.title('Match Win Percentage by Team and Season')
plt.xlabel('Home Team')
plt.ylabel('Season')
plt.show()
```

```python
win_percentage = data.groupby(['season', 'away_team']).apply(lambda x: (x['home_team'] == x['winner']).mean()).unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(win_percentage, cmap='YlGnBu', annot=True, fmt='.2f', cbar=True)
plt.title('Match Win Percentage by Team and Season')
plt.xlabel('Home Team')
plt.ylabel('Season')
plt.show()
```
